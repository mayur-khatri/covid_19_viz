```python
# READ DATA FROM CSV FILE
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
df = pd.read_csv('covid_19_india.csv') 
```

```python
df.sample(10)
```

```python
#df.loc[df['ConfirmedIndianNational'] == '-'] = 0
```

```python
df.dtypes
```

```python
df.replace('-', 0, inplace=True)
```

```python
#df.loc[df['ConfirmedForeignNational'] == '-'] = 0
```

```python
df.sample(10)
```

```python
df['ConfirmedIndianNational'] = df['ConfirmedIndianNational'].astype('int64')
```

```python
df['ConfirmedForeignNational'] = df['ConfirmedForeignNational'].astype('int64')
```

```python
df.dtypes
```

```python
# GROUP DATA FROM NAME OF STATE 
State = df.groupby(['State/UnionTerritory']).sum().reset_index()
State = State[1:]
```

```python
State
```

```python
State['ConfirmedIndianNational']
```

```python
State['State/UnionTerritory']
```

```python
# plot line graph between no. of Confirmed cases and state
x = State['State/UnionTerritory']
y = State['ConfirmedIndianNational']
z = State ['Cured']
z1= State ['Deaths']

plt.figure(figsize=(16,8)) # size of figure 16 * 8
plt.plot(x,y ,label = "Confirmed" , color = 'b', lw = 2 ) # plot line width 2
plt.title("TOTAL Confirmed CASES OF COVID-19 STATE WISE DATA INDIA", color = "b",fontweight="bold",fontsize=25) # title is a bold
plt.xticks(rotation=90,fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel("State Name",fontsize=16)
plt.ylabel("Number of Cases",fontsize=16)
plt.legend(loc= 'upper right',fontsize=16)
plt.show()
```

```python
df['Date']
```

```python
# GROUP DATA 
dateData = df.groupby(["Date"]).sum().reset_index()
# Convert that column into a datetime datatype
dateData['Date'] = pd.to_datetime(dateData['Date'], format= '%d/%m/%y')
#Sort by date
df_sorted = dateData.sort_values(by= "Date",ascending=True).set_index("Date")
df_sorted = df_sorted.groupby(['Date'])['Confirmed', 'Deaths','Cured'].sum().reset_index()
df_sorted
```

```python
# plot line graph between no. of cases and date
a = df_sorted.Date 
b = df_sorted.Confirmed
plt.figure(figsize=(16,8)) # size of figure 16 * 8
plt.plot(a,b)
plt.title("COVID-19 DATE WISE DATA INDIA",fontsize=16) # title and its font size 16
plt.xticks(rotation=60,fontsize=16) # x axis date rotate 60 degree and its font size is 16 
plt.xlabel("Date",fontsize=16) # x axis name is a Date its font size is 16 
plt.ylabel("Number of Cases",fontsize=16) # y axis name is a Number of Cases its font size is 16
plt.show()  # show the plot
```

```python
# plot line graph between no. of Deaths cases and state
plt.figure(figsize=(16,8)) # size of figure 16 * 8

plt.plot(x,z1 ,label = "Deaths" , color = 'r', lw = 2 )

plt.title("TOTAL Deaths CASES OF COVID-19 STATE WISE DATA INDIA", color = "b",fontweight="bold",fontsize=25)
plt.xticks(rotation=90,fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel("State Name",fontsize=16)
plt.ylabel("Number of Cases",fontsize=16)
plt.legend(loc= 'upper right',fontsize=16)
plt.show()
```

```python
# plot line graph between no. of cured cases and state
plt.figure(figsize=(16,8)) # size of figure 16 * 8

plt.plot(x,z ,label = "Cured" , color = 'g', lw = 2 )

plt.title("TOTAL CURED CASES OF COVID-19 STATE WISE DATA INDIA", color = "b",fontweight="bold",fontsize=25)
plt.xticks(rotation=90,fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel("State Name",fontsize=16)
plt.ylabel("Number of Cases",fontsize=16)
plt.legend(loc= 'upper right',fontsize=16)
plt.show()
```

```python
# plot line graph between no. of cured and Death cases and state
plt.figure(figsize=(16,8)) # size of figure 16 * 8

plt.plot(x,z ,label = "Cured" , color = 'g', lw = 2 )
plt.plot(x,z1 ,label = "Deaths" ,color ='r' ,linestyle='--',lw = 2 )

plt.title("TOTAL CURED and DEATHS CASES OF COVID-19 STATE WISE DATA INDIA", color = "b",fontweight="bold",fontsize=25)
plt.xticks(rotation=90,fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel("State Name",fontsize=16)
plt.ylabel("Number of Cases",fontsize=16)
plt.legend(loc= 'upper right',fontsize=16)
plt.show()
```

```python
# plot Pie graph between no. of Confirmed cases and state
from matplotlib import pyplot as plt
import numpy as np

fig = plt.figure(figsize=(11,11)) # size of figure 16 * 8
ax=fig.add_axes([0,0,1,1])
ax.axis('equal')
state = State['State/UnionTerritory']
cases = State['Confirmed']
ax.pie(cases, labels=state,autopct='%1.2f%%')
ax.set_title('Covid-19 INDIA', bbox={'facecolor':'0.8', 'pad':5})
plt.legend(state, loc="right",bbox_to_anchor=(1.4,0.5))

plt.show()
```

```python
# plot bar graph between no. of Confirmed cases and state with number
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
x = State['State/UnionTerritory']
y = State['Confirmed']
z = State ['Cured']
z1= State ['Deaths']
y_pos = np.arange(len(x))
fig = plt.figure(figsize=(20,8)) # size of figure 16 * 8
barplot = plt.bar(y_pos, y, align='center',color = list('rgbkymc'), alpha=0.5)
for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/8.0, yval, int(yval), va='bottom') #va: vertical alignment y positional argument
plt.xticks(y_pos, x, rotation=90,fontsize=16)
plt.xlabel('State Name',fontsize=16)
plt.ylabel('Number of Cases',fontsize=16)
plt.title("TOTAL CONFORMED CASES OF COVID-19 STATE WISE DATA INDIA", color = "b",fontweight="bold",fontsize=25)
plt.show()

```

```python
# plot Scatter graph between no. of Cured , Deaths cases and state
import matplotlib.pyplot as plt
x
fig=plt.figure(figsize=(12,4)) # size of figure 16 * 8

ax=fig.add_axes([0,0,1,1])
ax.scatter(x, z, color='b')
ax.scatter(x, z1, color='r', marker = '^')
plt.xticks(rotation=90,fontsize=16)
ax.set_xlabel('State Name ',fontsize=16)
ax.set_ylabel('No. cases',fontsize=16)
ax.grid(color='b', ls='-.', lw=0.25)
ax.set_title('Covid-19',fontsize=16)
ax.legend(labels=('Cured', 'Deaths'), loc='upper right')
plt.show()
```
